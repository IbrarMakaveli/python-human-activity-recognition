## 4. Exploratory Data Analysis

As per dataset description activites are split into two categories:

**Static**: Activites with no almost motion i.e. SITTING, STANDING, LAYING  
**Dynamic**: Activites with significant motion i.e. WALKING, WALKING_DOWNSTAIRS, WALKING_UPSTAIRS

### 4.1 Analyzing the Difference between two categories of Activites:

We can consider tBodyAccMag-mean() feature to differentiate among these two categories.

```python
facetgrid = sns.FacetGrid(df_train, hue='Activity', size=5,aspect=2)
facetgrid.map(sns.distplot,'tBodyAccMag-mean()', hist=False).add_legend()
plt.annotate("Static Activities", xy=(-.996,21), xytext=(-0.9, 23),arrowprops={'arrowstyle': '-', 'ls': 'dashed'})
plt.annotate("Static Activities", xy=(-.999,26), xytext=(-0.9, 23),arrowprops={'arrowstyle': '-', 'ls': 'dashed'})
plt.annotate("Static Activities", xy=(-0.985,12), xytext=(-0.9, 23),arrowprops={'arrowstyle': '-', 'ls': 'dashed'})
plt.annotate("Dynamic Activities", xy=(-0.2,3.25), xytext=(0.1, 9),arrowprops={'arrowstyle': '-', 'ls': 'dashed'})
plt.annotate("Dynamic Activities", xy=(0.1,2.18), xytext=(0.1, 9),arrowprops={'arrowstyle': '-', 'ls': 'dashed'})
plt.annotate("Dynamic Activities", xy=(-0.01,2.15), xytext=(0.1, 9),arrowprops={'arrowstyle': '-', 'ls': 'dashed'})
```

This density plot shows an important threshold of seperation between these two activites.  
    if tBodyAccMag-mean < -0.5:   
        'Static activities'  
    else if tBodyAccMag-mean > -0.5:  
        'Dynamic activities'  

Lets take a closer look at density plots of these two categories of activites

```python
plt.figure(figsize=(12,8))
plt.subplot(1,2,1)
plt.title("Static Activities")
sns.distplot(df_train[df_train["Activity"]=="SITTING"]['tBodyAccMag-mean()'],hist = False, label = 'Sitting')
sns.distplot(df_train[df_train["Activity"]=="STANDING"]['tBodyAccMag-mean()'],hist = False,label = 'Standing')
sns.distplot(df_train[df_train["Activity"]=="LAYING"]['tBodyAccMag-mean()'],hist = False, label = 'Laying')
plt.axis([-1.02, -0.5, 0, 35])
plt.subplot(1,2,2)
plt.title("Dynamic Activities")
sns.distplot(df_train[df_train["Activity"]=="WALKING"]['tBodyAccMag-mean()'],hist = False, label = 'Sitting')
sns.distplot(df_train[df_train["Activity"]=="WALKING_DOWNSTAIRS"]['tBodyAccMag-mean()'],hist = False,label = 'Standing')
sns.distplot(df_train[df_train["Activity"]=="WALKING_UPSTAIRS"]['tBodyAccMag-mean()'],hist = False, label = 'Laying')
```

We can also visualize this insight using box-plots

```python
plt.figure(figsize=(10,7))
sns.boxplot(x='Activity', y='tBodyAccMag-mean()',data=df_train, showfliers=False)
plt.ylabel('Body Acceleration Magnitude mean')
plt.title("Boxplot of tBodyAccMag-mean() column across various activities")
plt.axhline(y=-0.7, xmin=0.05,dashes=(3,3))
plt.axhline(y=0.020, xmin=0.35, dashes=(3,3))
plt.xticks(rotation=90)
```

1. Using box-plot we are getting the insight that:

## 5. Modeling

### 5.1 Importing classification models

```python
from sklearn. linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.svm import LinearSVC
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier

import warnings
warnings.filterwarnings("ignore")
```

### 5.2 Defining global reusable function

```python
def best_grid_search_results(model):
    print("Best estimator : ", model.best_estimator_)
    print("Best set of parameters : ", model.best_params_)
    print("Best score : ", model.best_score_)
    
def plot_confusion_matrix(y_test, y_true):
    cm = confusion_matrix(y_test,y_pred)
    df_cm = pd.DataFrame(cm, columns=np.unique(y_true), index = np.unique(y_true))
    df_cm.index.name = 'Actual'
    df_cm.columns.name = 'Predicted'
    plt.figure(figsize = (10,7))
    plt.ylim([1.5, -.5])
    sns.set(font_scale=1.4)#for label size
    sns.heatmap(df_cm, cmap="Blues", annot=True,annot_kws={"size": 12})# font size
```

### 5.3 Preparing Training and Test data for Machine Learning Models

```python
X_train = df_train.drop(['subject', 'Activity'], axis=1)
y_train = df_train.Activity
X_test = df_test.drop(['subject', 'Activity'], axis=1)
y_test = df_test.Activity
print('Training data size : ', X_train.shape)
print('Test data size : ', X_test.shape)
```

### 5.4 Comparison of Classifiers without Hyperparameter tuning

```python
classifiers = [
    LogisticRegression(),
    LinearSVC(tol=0.00005),
    SVC(),
    DecisionTreeClassifier(),
    KNeighborsClassifier(7), # 6 different labels 0-6
    RandomForestClassifier(200)
]

names = []
scores = []

for clf in classifiers:
    clf = clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    
    names.append(clf.__class__.__name__)
    scores.append(accuracy_score(y_pred, y_test))

score_df = pd.DataFrame({'Model': names, 'Score': scores}).set_index('Model')
score_df
```

```python
ax = score_df.plot.bar()
ax.set_xticklabels(score_df.index, rotation=45, fontsize=10)
```

Without hyperparameter tuning, we can see that Logistic regression has the highest accuracy. Now lets check which classifier performs best after hyperparameter tuning.

### 5.5 Comparison of Classifiers with Hyperparameter tuning

```python
# for storing the scores of classifiers after hyperparameter tuning
scores_tuned = []
```

**1. Logistic Regression**

```python
params = {'C':np.arange(10,61,10), 'penalty':['l2','l1']}
lr = LogisticRegression()
lr_tuned = GridSearchCV(lr, param_grid=params, cv=5)
lr_tuned.fit(X_train, y_train)
y_pred = lr_tuned.predict(X_test)

lr_accuracy = accuracy_score(y_test, y_pred)

print('LR Classifier Accuracy: ', lr_accuracy)
print('Classification Report: \n', classification_report(y_test, y_pred))
```

```python
scores_tuned.append(lr_tuned.best_score_)
scores_tuned
```

```python
best_grid_search_results(lr_tuned)
```

**2. Support Vector Machine**

- **Linear SVM**

```python
params = {'C':np.arange(1,12,2)}
lr_svm = LinearSVC(tol=0.00005)
lr_svm_tuned = GridSearchCV(lr_svm, param_grid=params)
lr_svm_tuned.fit(X_train, y_train)
y_pred = lr_svm_tuned.predict(X_test)

lr_svm_accuracy = accuracy_score(y_test, y_pred)


print('Linear SVM Classifier Accuracy: ', lr_svm_accuracy)
print('Classification Report: \n', classification_report(y_test, y_pred))
```

```python
scores_tuned.append(lr_svm_tuned.best_score_)
scores_tuned
```

```python
best_grid_search_results(lr_svm_tuned)
```

- **Kernel SVM**

```python
params = {'C':[2,4,8,16],'gamma': [0.125, 0.250, 0.5, 1]}
kernel_svm = SVC(kernel='rbf')
kernel_svm_tuned = GridSearchCV(kernel_svm, param_grid=params)
kernel_svm_tuned.fit(X_train, y_train)
y_pred = kernel_svm_tuned.predict(X_test)

kernel_svm_accuracy = accuracy_score(y_test, y_pred)

print('Kernel SVM Classifier Accuracy: ', kernel_svm_accuracy)
print('Classification Report: \n', classification_report(y_test, y_pred))
```

```python
scores_tuned.append(kernel_svm_tuned.best_score_)
scores_tuned
```

```python
best_grid_search_results(kernel_svm_tuned)
```

**3. Decision Tree Classifier**

```python
params = {'max_depth':np.arange(2,10,2)}
dtc = DecisionTreeClassifier()
dtc_tuned = GridSearchCV(dtc, param_grid=params)
dtc_tuned.fit(X_train, y_train)
y_pred = dtc_tuned.predict(X_test)

dtc_accuracy = accuracy_score(y_test, y_pred)

print('Decision Tree Classifier Accuracy: ', dtc_accuracy)
print('Classification Report: \n', classification_report(y_test, y_pred))
```

```python
scores_tuned.append(dtc_tuned.best_score_)
scores_tuned
```

```python
best_grid_search_results(dtc_tuned)
```

**4. K-Nearest Neighbor**

```python
params = {'n_neighbors': np.arange(1, 25)}
knn = KNeighborsClassifier()
knn_tuned = GridSearchCV(knn, param_grid=params, cv=5)
knn_tuned.fit(X_train, y_train)
y_pred = knn_tuned.predict(X_test)

knn_accuracy = accuracy_score(y_test, y_pred)

print('K-Nearest Neighbor Classifier Accuracy: ', knn_accuracy)
print('Classification Report: \n', classification_report(y_test, y_pred))
```

**5. Random Forests Classifier**

```python
params = {'n_estimators': np.arange(20,101,10), 'max_depth':np.arange(2,16,2)}
rf_classifier = RandomForestClassifier()
rf_classifier_tuned = GridSearchCV(rf_classifier, param_grid=params)
rf_classifier_tuned.fit(X_train, y_train)
y_pred = rf_classifier_tuned.predict(X_test)

rf_accuracy = accuracy_score(y_test, y_pred)

print('Random Forests Classifier Accuracy: ', rf_accuracy)
print('Classification Report: \n', classification_report(y_test, y_pred))
```
